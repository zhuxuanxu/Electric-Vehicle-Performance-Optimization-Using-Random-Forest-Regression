```
#COPY THIS CELL
def clean(df1, df2):
  df1[' HOUR'] = pd.to_datetime(df1[' HOUR'])
  df1['DT Elapsed'] = (df1.iloc[0:, 14] - df1.iat[1, 14])
  df1.drop(df1.index[[0,1]])
  df1['DT Elapsed'] = df1['DT Elapsed'].apply(lambda x: abs(x) if pd.notna(x) else None)
  df1.columns = df1.columns.str.replace(' ', '') 

  df2['TIME_UTC_STR'] = pd.to_datetime(df2['TIME_UTC_STR'])
  df2['Elapsed'] = df2.iloc[1:, 1] - df2.iat[0, 1]
  df2['AIR_TEMP'] = df2['AIR_TEMP'].astype(str)
  df2['TRACK_TEMP'] = df2['TRACK_TEMP'].astype(str)
  df2['WIND_SPEED'] = df2['WIND_SPEED'].astype(str)
  df2['AIR_TEMP'] = df2['AIR_TEMP'].str.replace(r'\D','.')
  df2['TRACK_TEMP'] = df2['TRACK_TEMP'].str.replace(r'\D','.')
  df2['WIND_SPEED'] = df2['WIND_SPEED'].str.replace(r'\D','.')

  df2['AIR_TEMP'] = df2['AIR_TEMP'].astype('float')
  df2['TRACK_TEMP'] = df2['TRACK_TEMP'].astype('float')
  df2['WIND_SPEED'] = df2['WIND_SPEED'].astype('float')
  return df1, df2
```

```
#COPY THIS CELL
def mergeweather(df1, df2):
  airtemplist = []
  tracktemplist = []
  windspeedlist = []
  winddirlist = []
  drivers = df1.NUMBER.unique()
  for j in range(len(drivers)):
    driverdf = df1[df1['NUMBER'] == drivers[j]].copy()
    idx = 0
    for i in range(len(driverdf)):
      check = driverdf.iloc[i,-1]
      selection = df2[(df2.iloc[0:, -1] - df2.iat[1, -1]) <= check].copy()
      idxlist = selection.index.tolist()
      selection = selection.iloc[idx-1:]
      avg = selection.mean()
      airtemp = avg['AIR_TEMP']
      tracktemp = avg['TRACK_TEMP']
      windspeed = avg['WIND_SPEED']
      winddir = avg['WIND_DIRECTION']
      airtemplist.append(airtemp)
      tracktemplist.append(tracktemp)
      windspeedlist.append(windspeed)
      winddirlist.append(winddir)
      idx = int(idxlist[-1])


  weatherdict = {'AIR_TEMP':airtemplist,'TRACK_TEMP':tracktemplist,'WIND_SPEED':windspeedlist,'WIND_DIRECTION':winddirlist}
  weatherdf = pd.DataFrame(weatherdict)
  merge = pd.concat([df1,weatherdf],axis=1,ignore_index=False,join='outer')
  df1 = merge.copy()
  return df1
```

```
#COPY THIS CELL
def dropcol(df):
  df.drop(labels=['DRIVER_NUMBER', 'CROSSING_FINISH_LINE_IN_PIT','S1','S1_IMPROVEMENT','S2','S2_IMPROVEMENT','S3','S3_IMPROVEMENT','S1_LARGE','S2_LARGE','S3_LARGE','TOP_SPEED', 'DTElapsed','Unnamed:25'], axis=1, inplace=True)
  return df
```
