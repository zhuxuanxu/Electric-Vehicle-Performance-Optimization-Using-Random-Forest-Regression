```
# 2016-2017 Q
B16Q = pd.read_csv('/content/drive/Shared drives/Eco-Cheetah F20 Case Comp/Code/Cleaned CSVs/B16Q.csv')
HK16Q = pd.read_csv('/content/drive/Shared drives/Eco-Cheetah F20 Case Comp/Code/Cleaned CSVs/HK16Q.csv')
MA16Q = pd.read_csv('/content/drive/Shared drives/Eco-Cheetah F20 Case Comp/Code/Cleaned CSVs/MA16Q.csv')
MC16Q = pd.read_csv('/content/drive/Shared drives/Eco-Cheetah F20 Case Comp/Code/Cleaned CSVs/MC16Q.csv')
NY16Q = pd.read_csv('/content/drive/Shared drives/Eco-Cheetah F20 Case Comp/Code/Cleaned CSVs/NY16Q.csv')
PR16Q = pd.read_csv('/content/drive/Shared drives/Eco-Cheetah F20 Case Comp/Code/Cleaned CSVs/PR16Q.csv')
```

```
concat_16Q = pd.concat([B16Q,HK16Q,MA16Q,MC16Q,NY16Q,PR16Q], axis=0, ignore_index = True)
```

```
# drop rows which don't have weather data
concat_16Q = concat_16Q[concat_16Q['TRACK_TEMP'].notna()]
concat_16Q
```

```
# change lap time to seconds
concat_16Q['LAP_TIME'] = concat_16Q['LAP_TIME'].map(lambda x:(float(x.split(':')[0])*60 + float(x.split(':')[1])))
concat_16Q
```

```
### ignore human intervention factors like crash possibilities, or safety car's impact on the race timing
# delete the rows which lap time exceeds 80 seconds
concat_16Q = concat_16Q[concat_16Q['LAP_TIME'] <= 80]
concat_16Q
```

```
concat_16Q['ELAPSED'] = concat_16Q['ELAPSED'].map(lambda x:(float(x.split(':')[0])*60 + float(x.split(':')[1])))
```

```
X=concat_16Q[['LAP_IMPROVEMENT','AIR_TEMP','TRACK_TEMP','WIND_SPEED','ELAPSED']] #,
y=concat_16Q['LAP_TIME']
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.2)
```

```
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```
model = linear_model.LinearRegression()
model.fit(X_train, Y_train)
```

```
Y_pred_train = model.predict(X_train)
```

```
print('Coefficients:', model.coef_)
print('Intercept:', model.intercept_)
print('Mean squared error (MSE): %.2f'
      % mean_squared_error(Y_train, Y_pred_train))
print('Coefficient of determination (R^2): %.2f'
      % r2_score(Y_train, Y_pred_train))

#disp = plot_confusion_matrix(classifier, X_test, Y_test)
```

```
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
y_pred = model.predict(X_test)
#print(confusion_matrix(Y_test.astype(int),y_pred.astype(int)))
#print(classification_report(Y_test,y_pred))
print(accuracy_score(Y_test.astype(int), y_pred.astype(int)))
#print(y_pred)
#print(Y_test)
```

```
# keep the columns that need for regression
concat_16Q = concat_16Q[['LAP_TIME','TRACK_TEMP','AIR_TEMP','WIND_SPEED','WIND_DIRECTION']]
concat_16Q
```
