```
import pandas as pd
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
```

```
df1 = pd.read_csv('/content/drive/Shared drives/Eco-Cheetah F20 Case Comp/Code/Final CSVs/2016 Qualifying Race.csv')
df2 = pd.read_csv('/content/drive/Shared drives/Eco-Cheetah F20 Case Comp/Code/Final CSVs/2017 Qualifying Race.csv')
df3 = pd.read_csv('/content/drive/Shared drives/Eco-Cheetah F20 Case Comp/Code/Final CSVs/2016 Super Pole Race.csv')
df4 = pd.read_csv('/content/drive/Shared drives/Eco-Cheetah F20 Case Comp/Code/Final CSVs/2017 Super Pole Race.csv')
df5 = pd.read_csv('/content/drive/Shared drives/Eco-Cheetah F20 Case Comp/Code/Final CSVs/2016 Final Race.csv')
df5 = pd.read_csv('/content/drive/Shared drives/Eco-Cheetah F20 Case Comp/Code/Final CSVs/2017 Final Race.csv')
df1.head()
```

```
df1 = df1[df1['TRACK_TEMP'].notna()]
```

```
dataset = df1.values
```

```
X = dataset[:,15:19].astype(float)
Y = dataset[:,4]
```

```
min_max_scaler = preprocessing.MinMaxScaler()
X_scale = min_max_scaler.fit_transform(X)
```

```
X_train, X_val_and_test, Y_train, Y_val_and_test = train_test_split(X_scale, Y, test_size=0.3)
```

```
X_val, X_test, Y_val, Y_test = train_test_split(X_val_and_test, Y_val_and_test, test_size=0.5)
```

```
print(X_train.shape, X_val.shape, X_test.shape, Y_train.shape, Y_val.shape, Y_test.shape)
```

```
model = Sequential([
    Dense(32, activation='relu', input_shape=(4,)),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid'),
])
```

```
model.compile(optimizer='sgd',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```
hist = model.fit(X_train, Y_train,
          batch_size=32, epochs=100,
          validation_data=(X_val, Y_val))
```
